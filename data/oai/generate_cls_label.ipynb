```python
import os
import pandas as pd
import numpy as np
```

```python
folder_path = './Package_1197300/'
package_info = pd.read_csv(folder_path + 'kxr_sq_bu00.txt', sep = "|")
```

```python
pathnames =[]
for root, dirs, files in os.walk(folder_path+'results/'):
    for filename in files:
        if 'jpg' not in filename:
            pathnames += [os.path.join(root, filename)]
pathnames = pathnames[3:]
```

```python
pathname_e = [x for x in pathnames if x.find('0.C.2')==-1]
```

```python
unique_pathnames = []
for x in pathname_e:
    if x not in unique_pathnames:
        unique_pathnames.append(x)
print(len(unique_pathnames))
```

```python
df = pd.DataFrame(columns=['ID','DICOM_PATH','R_KL','L_KL','R_JSN_M', 'R_JSN_L', 'L_JSN_M', 'L_JSN_L'])
col_now=0
for i in range(len(pathnames)):
    pathname = pathnames[i]
    if not pathname.split("/")[-1]=='.DS_Store':
        ID = pathname.split("/")[5]
        col1 = package_info['ID']==int(ID)
        col2 = package_info['READPRJ']==15
        KL=[-1,-1]
        JSN_M = [-1,-1]
        JSN_L = [-1,-1]
        get = -1
        for side in [1,2]:
            col3 = package_info['SIDE']==side
            col_final=package_info[col1&col2&col3]
            if len(col_final>0):
                KL[side-1] = col_final['V00XRKL'].iloc[0]
                JSN_M[side-1] = col_final['V00XRJSM'].iloc[0]
                JSN_L[side-1] = col_final['V00XRJSL'].iloc[0]
                get = 1
            else:
                print(pathname)
        if get == 1:
            df.loc[col_now] = [ID,pathname,KL[0],KL[1],JSN_M[0],JSN_M[1],JSN_L[0],JSN_L[1]]
            col_now+=1
df.to_csv(r'oai_0E1.csv',index = False, header=True)
```
